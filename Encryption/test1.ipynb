```python
def get_data(filename):

    with open(filename, "r", encoding= "UTF-8") as f:
        data = f.readlines()
    return data

file = "key.txt"
data = get_data(file)
print(data)
```

```python
def encodingFuncGen(key):
    def encodingFunc(char):
        return char + key
    return encodingFunc
```

```python
def encode(data, encodingFunc):
    #encode the list, data\n",
    datain = []
    for row in data:
        rowin = ""
        for char in row:
            rowin += chr(encodingFunc(ord(char)))
        datain.append(rowin)
    return datain
print(data)
enFunc = encodingFuncGen(1)
print(encode(data, enFunc))
```

```python
def write_data(filename, data):
    with open(filename, "w") as f:
        for element in data:
            f.write(element)
        
write_data("testfile.txt", data)
```

```python
def encode_file(filename, Key):
    #Generate encoding function
    enFunc= encodingFuncGen(Key)
    #Get data from file
    data = get_data(filename)
    #Encode data
    datain = []
    for row in data:
        rowin = ""
        for char in row:
            rowin += chr(enFunc(ord(char)))
        datain.append(rowin)
    data = datain
    #Write data to file
    newfilename = filename + ".enc"
    write_data(newfilename, data)

encode_file("testkey.txt", 1)   
```

```python
def decode_file(filename, Key):
    #Generate decoding function
    enFunc= encodingFuncGen(Key)
    #Get data from file
    data = get_data(filename)
    #Encode data
    dataout = []
    for row in data:
        rowin = ""
        for char in row:
            rowin += chr(enFunc(ord(char)))
        dataout.append(rowin)
    data = dataout
    #Write data to file
    newfilename = filename[:-4] + ".dec"
    write_data(newfilename, data)
decode_file("testkey.txt.enc", -1)
```

```python
def main():
    filename = input("Enter filename: ")
    choice = input("Would you like to encode or decode a file? (e/d): ")
    if choice == "e":
        key = int(input("Enter key: "))
        encode_file(filename, key)
        print("File encoded")
    elif choice == "d":
        key = int(input("Enter key: "))
        decode_file(filename, key)
        print("File decoded")
    else:
        print("Invalid choice")
        main()
```
